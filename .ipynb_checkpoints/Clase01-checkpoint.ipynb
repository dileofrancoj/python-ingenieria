## Métodos y funciones asociados a arrays


## Condicionales

Un condicional es una sentencia que me permite evaluar una expresion / compararla con otra cosa y en función de su resultado tomar decisiones (ramificar el programa)

```python
    a == b # igualdad (2 iguales comparan, un igual asigna)
    a != b # distinto
    a > b # a mayor a b 
    a < b 
    a >= b
    ...
    etc
```

````python
numbers = [25.0,73, 13, 100, 200]
````

````python
numA = numbers[0]
numB = numbers[-1]
````

````python
if numA < numB:
    print(f"{numA} es más chico que {numB}")
````
