# Matplotlib

## Algunas utilidades

$$
\textbf{Funciones principales de Matplotlib (pyplot)}
$$

$$
\begin{aligned}
1.\ & \texttt{plt.plot(x, y)} && \rightarrow \text{Gráfico de líneas} \\
2.\ & \texttt{plt.scatter(x, y)} && \rightarrow \text{Gráfico de dispersión (puntos)} \\
3.\ & \texttt{plt.hist(data, bins)} && \rightarrow \text{Histograma} \\
4.\ & \texttt{plt.bar(categorías, valores)} && \rightarrow \text{Gráfico de barras} \\
5.\ & \texttt{plt.imshow(Z)} && \rightarrow \text{Mapa de calor (matriz como imagen)} \\
6.\ & \texttt{plt.contour(X, Y, Z)} && \rightarrow \text{Curvas de nivel (funciones 2D)} \\
7.\ & \texttt{plt.title("Título")} && \rightarrow \text{Título del gráfico} \\
8.\ & \texttt{plt.xlabel("Etiqueta X")} && \rightarrow \text{Etiqueta del eje X} \\
9.\ & \texttt{plt.ylabel("Etiqueta Y")} && \rightarrow \text{Etiqueta del eje Y} \\
10.\ & \texttt{plt.legend()} && \rightarrow \text{Mostrar leyenda de curvas o series} \\
11.\ & \texttt{plt.grid(True)} && \rightarrow \text{Agregar cuadrícula} \\
12.\ & \texttt{plt.xlim(a,b),\ plt.ylim(c,d)} && \rightarrow \text{Límites de ejes} \\
13.\ & \texttt{plt.annotate("texto", xy=(x,y))} && \rightarrow \text{Anotar un punto específico} \\
14.\ & \texttt{plt.subplots(n,m)} && \rightarrow \text{Crear múltiples subgráficos} \\
15.\ & \texttt{plt.show()} && \rightarrow \text{Mostrar el gráfico en pantalla}
16.\ & \texttt{plt.quiver(x,y, U,V)} && \rightarrow \text{Vector} \\
\end{aligned}
$$


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# en caso de no tener las librerias y estar en local (jupyter notebook) pip install numpy
# en caso de estar en collab no hay que instalar nada, viene todo por defecto
```

```python
x = np.linspace(0,10,5)

y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
x2 = np.linspace(0,10,200)
y2 = np.sin(x2)
```

```python
plt.plot(x2,y2)
plt.title("$\sin(x)$")
plt.xlabel("x")
plt.grid()
plt.ylabel("y")
plt.show()
```

```python
x = np.linspace(0.1,10,150)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)
```

```python
plt.plot(x,y1,color='red', label="$\sin(x)$")
plt.plot(x,y2, color='green', label="$\cos(x)$")
plt.plot(x,y3, color='orange', label="$e^x$", linestyle="--")
plt.title("funciones")
plt.grid()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = np.linspace(-10,10,100)
y1 = np.tan(x)
y2 = np.cosh(x)
y3 = np.log2(x+11)
y4 = np.power(2,x)
```

```python
y = np.random.rand(len(x))*50
y5 = 1.2*x
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,6))
axs[0][0].plot(x,y1,color='green')
axs[0][0].scatter(x,y,color='purple', marker='o')
axs[0][0].set_title("tan")

axs[0][1].plot(x,y2,color='red')
axs[0][1].plot(x,x**2,color='purple')
axs[0][1].set_title("cosh(x)")

axs[0][1].plot(x,y2,color='red')
axs[0][1].set_title("cosh(x)")

axs[1][0].plot(x,y3,color='yellow')
axs[1][0].set_title("log2(x)")

axs[1][1].scatter(x,y,color='red', linewidth=1)
axs[1][1].scatter(x,y5,color='green')
axs[1][1].set_title("y(x)")
```

## Ejercicios

Comparar las funciones $\sin(x)$ y $\cos(x)$ en el mismo gráfico en el intervalo $[0,2\pi]$

```python
x_2 = np.linspace(0,2*np.pi,100) # generamos un intervalo de 0 a 2 con 100 elementos
y_1 = np.sin(x_2)
y_2 = np.cos(x_2)
```

```python
plt.plot(x_2,y_1, color='red', label="$f(x) = \cos(x)$")
plt.plot(x_2,y_2, color='purple', label="$f(x) = \sin(x)$", linestyle="--")
plt.title("Comparativa de funciones")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, alpha=0.4)
```

# Ejercicio 2
Dispersión con ruido: Generar datos sintéticos de una recta con un cierto ruido $\epsilon$

$y = 2x + 1 + \epsilon$

```python
elements = 100
epsilon = np.random.normal(size=elements)
```

```python
x_3 = np.linspace(0,10, elements)
y_3 = 2*x_3 + 1 # x_3 tiene 100 elementos , epsilon tiene 100 elementos
y_3_noise = y_3 + epsilon
```

```python
plt.scatter(x_3, y_3_noise, color='red', alpha=0.3)
plt.plot(x_3,y_3,color='blue')
plt.xlabel("x")
plt.xlabel("y")
plt.title("Dispersión normal y recta")
plt.grid(True, alpha=0.5)
plt.show()
```
