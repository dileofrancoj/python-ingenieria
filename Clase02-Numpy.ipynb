Antes de arrancar, vamos a aclarar algunos conceptos

# Vectores
Lo podemos atacar desde el punto de vista informatico que son elementos contiguos en la memoria, desde el punto de vista físico que es un segmento orientado de recta o bien desde el matemático que es: un vector es un elemento de un espacio vectorial (cerrado bajo sumas entre elemento de ese espacio y cerrado bajo multiplicaciones por un escalar) en este caso vamos a estar trabajando con el típico espacio vectorial (R^n, + , R , *) que es un EV definido sobre el cuerpo de los reales.

Un polinomio es un elemento de un EV.

# Matrices
Son colecciones de numeros (pueden ser reales, complejos, etc) es un arreglo de filas y columnas (pueden ser rectangulares, cuadradas) 
En una matriz se puede condensar información (pueden representar TL's, sistemas de ecuaciones, etc) aa priori fue nomenclatura pero rápidamente se le dio una estructura a esto.

# Tensores

Un tensor es la generalización de lo que acabamos de explicar. Particularmente un escalar es un tensor de orden 0, un vector es un tensor de orden 1, etc. Los tensores se usan mucho en física y matemáticas por ejemplo, para representar la métrica, el tensor de tensiones en mecánica, el tensor energía-momento en relatividad, etc.

En machine learning e IA podemos ver a los tensores como una matriz multidimensional que sirve para guardar información en un espacio de m1xm2xm3x...mn características (ejemplo imagen rgb)
