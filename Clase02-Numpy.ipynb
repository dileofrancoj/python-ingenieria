### Suma de matrices

$
A + B
$

### Producto elemento a elemento

$C = A \odot B$


### Producto matricial

Crea una matriz $2 \times 3$ y otra $3 \times 2$. Haz el producto matricial y verifica el tamaño del resultado.  
    
$
    C = A \times B
$

### Matriz inversa

Calcula la inversa de la siguiente matriz:
$
    A = 
    \begin{bmatrix}
    4 & 7 \\
    2 & 6
    \end{bmatrix}
$

### Verificación de inversa

Multiplica la matriz anterior por su inversa y demuestra que se obtiene la identidad:

$
    A \cdot A^{-1} = I
$
