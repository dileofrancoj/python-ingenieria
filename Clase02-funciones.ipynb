```python
def contar_numeros_pares(numeros):
    cantidad_numeros_pares = 0
    cantidad_numeros_impares = 0
    for numero in numeros:
        if numero%2 == 0:
            # el numero es par... --> acumulador
            cantidad_numeros_pares += 1 # cantidad_numeros_pares = cantidad_numeros_pares + 1
        else:
            cantidad_numeros_impares += 1
    print("Cantidad impares:", cantidad_numeros_pares)
    print("Cantidad impares:", cantidad_numeros_impares)
```

```python
contar_numeros_pares([2,3,8,10,11])
```
