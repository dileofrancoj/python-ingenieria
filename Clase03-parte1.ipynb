### Transformación lineal

Considera la matriz de rotación en el plano de ángulo $\theta = \frac{\pi}{2}$:

$
    R = 
    \begin{bmatrix}
    \cos \theta & -\sin \theta \\
    \sin \theta & \cos \theta
    \end{bmatrix}
    $
    Aplica la transformación a los vectores
    $
    v_1 = 
    \begin{bmatrix}
    1 \\
    0
    \end{bmatrix}, \quad
    v_2 = 
    \begin{bmatrix}
    0 \\
    1
    \end{bmatrix}
$
    y representa gráficamente el resultado.



```python
import numpy as np
```

```python
X = np.array([[1,1,1],[2,-1,3],[1,4,-1]])
b = np.array([6,14,2])
x1,x2,x3 = np.linalg.solve(X,b) # destructurar
```

```python
print(f"{x1},{x2},{x3}")
```

```python
np.linalg.det(X)
```

```python
np.linalg.inv(X)
```
